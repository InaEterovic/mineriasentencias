```python
import pymongo
import nltk
```

```python
client = pymongo.MongoClient("mongodb://localhost:27017/")
sentenceDB  = client["Sentencias"]
wordCollection =  sentenceDB["WordTokens"]
```

```python
x = wordCollection.find_one()
wordList = x['wordTokenList']
```

```python
from nltk import SnowballStemmer
spanishstemmer=SnowballStemmer('spanish')
def stemmingList(words):
    stems = [spanishstemmer.stem(word) for word in words]
    return stems
```

```python
stemmingList(wordList)
```
