```python
import pymongo
import pandas as pd
```

```python
client = pymongo.MongoClient("mongodb://localhost:27017/")
sentenceDB  = client["Sentencias"]
wordCollection = sentenceDB["WordTokens"]
```

```python
data = wordCollection.find()
df = pd.DataFrame(list(data))
```

```python
df.dtypes
```

```python
len(df)
```

```python
df.head()
```

```python
tokens = []
```

```python
c=0
for wordList in df.wordTokenList:
    tokens = tokens + wordList
    c=c+1
    if c % 100 == 0:
        print(c)
```

```python
from collections import Counter
```

```python
STOP_WORDS= ['“','”','el','que','por','a','nº','la','en','de']
cleanTokens = [w for w in tokens if w not in STOP_WORDS]
```

```python
counter = Counter(cleanTokens)
counter.most_common(100)
```

```python
wordfrequencies="".join(str(counter.most_common(100))).strip('[]')
wordfrequencies
```
